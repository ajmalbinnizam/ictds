```python
from sqlalchemy import create_engine  # step1
import pandas as pd   # step1
engine = create_engine("sqlite:///Chinook.sqlite")   # step2
conn = engine.connect()  # step3
query = "select * from Customer"
rs = conn.execute(query)  # 4 
df = pd.DataFrame(rs.fetchall()) # step5
conn.close()  # step 6
```

```python
# using context manager
```

```python
engine = create_engine("sqlite:///Chinook.sqlite")
with engine.connect() as con:
    rs = con.execute("SELECT * FROM Genre")
    df_genre = pd.DataFrame(rs.fetchall())
    df_genre.columns = rs.keys()
```

```python
df_genre.head()   # 3 gb
```

```python
engine = create_engine("sqlite:///Chinook.sqlite")
query = "select a.Name as TrackName, a.composer, b.Name as GenreName FROM Track as a  INNER JOIN Genre as b ON a.GenreId= b.GenreId"

with engine.connect() as con:
    rs = con.execute(query)
    df_join = pd.DataFrame(rs.fetchall())
    df_join.columns = rs.keys()
```
