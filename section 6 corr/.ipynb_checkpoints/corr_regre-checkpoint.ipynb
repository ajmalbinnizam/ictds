```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_excel('tip-amount.xlsx')
```

```python
data.head()
```

```python
corrmatrix = data.corr()
plt.subplots(figsize=(10,4))
sns.heatmap(corrmatrix,vmin=-.2,vmax=.9,annot=True,linewidths=0.2,cmap = 'YlGnBu')
```

```python
data.columns
```

```python
y = data['Observed tip amount(yi)']
#high correlation meal and observed tio amount ther are not related
x = data.drop(['Meal','Observed tip amount(yi)'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state = 42,test_size = 0.25)
```

```python
from sklearn import linear_model
lr = linear_model.LinearRegression()
model = lr.fit(x_train,y_train)
```

```python
predictions = model.predict(x_test)
```

```python
print(y_test)
```

```python
print(predictions)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print('mse is', mean_squared_error(y_test, predictions))
```

```python
for col in data.columns.values:
    print('number uniques values {} : {}'.format(col,data[col].nunique() ))
```

```python
sns.catplot(x= 'quality',data=data,kind='count')
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data.corr(), color='k', annot=True)
```

```python
plt.figure(figsize=(10,15))
for i, col in enumerate(list(data.columns.values)):
    plt.subplot(4,3,i+1)
    data.boxplot(col)
    plt.grid()
    plt.tight_layout()
```

```python
plt.figure(figsize=(20,16))
for i, col in enumerate(data.columns.values):
    plt.subplot(4,3,i+1)
    sns.distplot(data[col],color='r',kde=True,label='data')
    plt.grid()
    plt.legend(loc='upper right')
    plt.tight_layout()
```

```python
sns.pairplot(data=data, kind='scatter', diag_kind='kde')
```

```python
freqgraph = data.select_dtypes(include=['int32', 'int64', 'float'])
freqgraph.hist(figsize=(20,15))
plt.show()
```

```python
corrmatrix = data.corr()
plt.subplots(figsize=(20,8))
sns.heatmap(corrmatrix, vmin=-.7,vmax=0.7,annot=True,linewidth=0.2)
```

```python
data.columns
```

```python
data1 = data.drop(['fixed acidity','density', 'free sulfur dioxide'],axis = 1)
```

```python
data1.columns
```

```python
y = data['quality']
x = data1.drop(['quality'],axis=1)
```

```python
x.head()
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x,y,random_state=42,test_size=0.25)
```

```python
from sklearn import linear_model
lr = linear_model.LinearRegression()
model= lr.fit(x_train,y_train)
prediction= model.predict(x_test)
```

```python
print(y_test)
```

```python
print(prediction)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print('mse is', mean_squared_error(y_test,prediction))
```
