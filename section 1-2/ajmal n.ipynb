<h3>HISTOGRAM<h3>

```python
plt.figure(figsize=(6,4))
plt.hist(data['SL'],color='red', rwidth=0.9,density=True,cumulative=True)
plt.title("sepal length",fontsize=16)

plt.xlabel('sepal length')
plt.ylabel('frequency')

plt.xticks()
plt.yticks()
```

```python
plt.hist(data['SL'])
plt.hist(data['SW'])
plt.xlabel('sepal length sepal width')
```

```python
plt.hist(data['PL'])
plt.hist(data['PW'])
plt.xlabel('petal length      petal width')
```

```python
plt.hist(data['Classification'])
```

<h3>KDE<h3>

```python
data[['SL','SW','PL','PW']].plot.kde()
```

<h3>SCATTER PLOT<h3>

```python
plt.figure(figsize=(6,4))
plt.scatter(data['SL'],data['SW'],s=12, c = 'red',marker = '^')
plt.title("plot of SL vs SW ",fontsize=16)

plt.xlabel('sepal length')
plt.ylabel('sepal width')
```

<h3>PLOT GRAPH<h3>

```python
data1 = pd.DataFrame(data,columns = ['SL'])

data2 = pd.DataFrame(data,columns = ['SW'])
```

```python
plt.plot(data1,data2)
plt.title('sepal length vs sepal width')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
```
