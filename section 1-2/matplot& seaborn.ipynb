# histogram


```python
data = pd.read_csv('auto-mpg.csv')
```

```python
data
```

```python
data.columns
```

```python
plt.figure()
plt.hist(data['mpg'])
plt.title('distri')
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(8,8)) #size of the hist
plt.hist(data['mpg'])
plt.title('distri')
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(8,8))
plt.hist(data['mpg'],color='red')
plt.title("distribution of mfg",fontsize=16)
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(8,8))
plt.hist(data['mpg'],color='red')
plt.title("distribution of mfg",fontsize=16)

plt.xlabel('km/hr')
plt.ylabel('frequency')

plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(8,8))
plt.hist(data['mpg'],color='red', rwidth=0.9)
plt.title("distribution of mfg",fontsize=16)

plt.xlabel('km/hr')
plt.ylabel('frequency')

plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(8,8))
plt.hist(data['mpg'],color='red', rwidth=0.9,density= True)
plt.title("distribution of mfg",fontsize=16)

plt.xlabel('km/hr')
plt.ylabel('frequency')

plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(8,8))
plt.hist(data['mpg'],color='red', rwidth=0.9,density=True,cumulative=True)
plt.title("distribution of mfg",fontsize=16)

plt.xlabel('km/hr')
plt.ylabel('frequency')

plt.xticks()
plt.yticks()
```

# scatter plot

```python
plt.figure(figsize=(6,4))
plt.scatter(data1['horsepower'],data1['mpg'],s=10)
plt.title("plot of hp vs mpg ",fontsize=16)
plt.xlabel('km/hr')
plt.ylabel('frequency')
```

```python
plt.figure(figsize=(6,4))
plt.scatter(data1['horsepower'],data1['mpg'],s=10, color = 'violet')
plt.title("plot of hp vs mpg ",fontsize=16)
plt.xlabel('km/hr')
plt.ylabel('frequency')
```

```python
plt.figure(figsize=(6,4))
plt.scatter(data1['horsepower'],data1['mpg'],s=12, color = 'red',marker = '^')
plt.title("plot of hp vs mpg ",fontsize=16)

plt.xlabel('km/hr')
plt.ylabel('frequency')
```

```python
plt.figure(figsize=(6,4))
plt.scatter(data1['displacement'],data1['mpg'],s=12, c = 'red',marker = '^')
plt.title("plot of hp vs mpg ",fontsize=16)

plt.xlabel('disp')
plt.ylabel('km/hr')
```
